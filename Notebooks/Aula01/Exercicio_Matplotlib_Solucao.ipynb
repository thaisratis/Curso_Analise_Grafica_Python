```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../../Datasets/german_credit_data.csv')
data
```

```python
# Gráfico de Linha: Variação de idades ao longo dos registros
plt.figure(figsize=(10, 6))
plt.plot(data['Age'])
plt.title('Variação de Idades')
plt.xlabel('Registro')
plt.ylabel('Idade')
plt.show()
```

```python
data['Sex'].value_counts()
```

```python
# Gráfico de Barras: Contagem de Pessoas por Sexo

sex_counts = data['Sex'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(sex_counts.index, sex_counts.values, color='skyblue')
plt.title('Contagem de Pessoas por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.show()
```

```python
# Gráfico de Pizza: Distribuição dos propósitos
plt.figure(figsize=(8, 8))
plt.pie(data['Purpose'].value_counts(), labels=data['Purpose'].unique(), autopct='%1.1f%%', startangle=140, textprops={'color':"white"})
plt.title('Distribuição dos propósitos', color='white')
plt.show()
```

```python
# Histograma: Distribuição de Idades
plt.figure(figsize=(10, 6))
plt.hist(data['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuição de Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de Dispersão: Idade vs. Valor do Crédito
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Credit amount'], alpha=0.5)
plt.title('Gráfico de Dispersão: Idade vs. Valor do Crédito')
plt.xlabel('Idade')
plt.ylabel('Valor do Crédito')
plt.show()
```

```python
# Gráfico de Boxplot: Distribuição de Idade por Sexo
plt.figure(figsize=(8, 6))
plt.boxplot([data[data['Sex'] == 'male']['Age'], data[data['Sex'] == 'female']['Age']], labels=['Masculino', 'Feminino'])
plt.title('Comparação de Idade por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Idade')
plt.show()
```

```python
# Gráfico de dispersão para mostrar a relação entre idade e quantidade de crédito (campo "Age" vs. "Credit amount"), onde as cores representam o sexo e o tamanho dos pontos varia conforme o propósito

# Definindo as cores para o sexo
colors = {'male': 'blue', 'female': 'red'}

# Definindo o tamanho dos pontos com base no propósito
sizes = data['Purpose'].map({'radio/TV': 50, 'education': 100, 'furniture/equipment': 150, 'car': 200})

# Criando o gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Credit amount'], c=data['Sex'].map(colors), s=sizes, alpha=0.6)
plt.xlabel('Idade')
plt.ylabel('Quantidade de Crédito')
plt.title('Relação entre Idade e Crédito com Sexo e Propósito')
plt.show()
```

```python
# Gráfico de linha mostrando a variação da quantidade de crédito em função da duração do crédito, separando por sexo.

# Filtrando os dados por sexo
male_data = data[data['Sex'] == 'male']
female_data = data[data['Sex'] == 'female']

# Criando o gráfico de linha

plt.figure(figsize=(10, 6))
plt.plot(male_data['Duration'], male_data['Credit amount'], label='Masculino', color='blue')
plt.plot(female_data['Duration'], female_data['Credit amount'], label='Feminino', color='pink')

plt.xlabel('Duração do Crédito (meses)')
plt.ylabel('Quantidade de Crédito')
plt.title('Variação da Quantidade de Crédito em Função da Duração')
plt.legend()
plt.show()
```

```python
# Histogramas comparando a distribuição de idades entre masculino e feminino

# Filtrando os dados por sexo
male_ages = data[data['Sex'] == 'male']['Age']
female_ages = data[data['Sex'] == 'female']['Age']

# Criando os histogramas
plt.hist(male_ages, bins=10, alpha=0.5, label='Masculino', color='blue', edgecolor='black')
plt.hist(female_ages, bins=10, alpha=0.5, label='Feminino', color='pink', edgecolor='black')


plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.title('Distribuição de Idade por Sexo')
plt.legend()
plt.show()
```
