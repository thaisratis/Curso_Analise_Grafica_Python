```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('../../Datasets/german_credit_data.csv')
data
```

```python
# Gráfico de Linha: Variação de idades ao longo dos registros
plt.figure(figsize=(10, 6))
sns.lineplot(data=data['Age'])
plt.title('Variação de Idades')
plt.xlabel('Registro')
plt.ylabel('Idade')
plt.show()
```

```python
# Gráfico de contagem: Contagem de Pessoas por Sexo

plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Sex')
plt.title('Contagem de Pessoas por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.show()
```

```python
# Gráfico de Barras usando barplot
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x='Sex', y='Age', estimator=len) 
# O parâmetro estimator=len especifica que queremos contar o número de ocorrências em vez de calcular a média.
plt.title('Contagem de Pessoas por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.show()
```

```python
# Gráfico de Barras (Barplot) usando barplot
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x='Sex', y='Credit amount', ci=None)
# O parâmetro ci=None desativa a exibição dos intervalos de confiança.
plt.title('Média do Valor do Crédito por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Média do Valor do Crédito')
plt.show()
```

```python
# Histograma: Distribuição de Idades
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuição de Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de Dispersão: Idade vs. Valor do Crédito
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Age', y='Credit amount', alpha=0.5)
plt.title('Gráfico de Dispersão: Idade vs. Valor do Crédito')
plt.xlabel('Idade')
plt.ylabel('Valor do Crédito')
plt.show()
```

```python
# Gráfico de Boxplot: Distribuição de Idade por Sexo
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Sex', y='Age')
plt.title('Comparação de Idade por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Idade')
plt.show()
```

```python
# Calcular a matriz de correlação
correlation_matrix = data.corr()

# Gráfico de Mapa de Calor (Heatmap)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Mapa de Calor: Matriz de Correlação')
plt.show()
```

```python
# Gráfico de dispersão: Idade vs. Quantidade de Crédito
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=data,
    x="Age",
    y="Credit amount",
    hue="Sex",              # cores por sexo
    size="Purpose",         # tamanho por propósito
    sizes=(40, 200),        # define a faixa de tamanhos
    alpha=0.7
)
plt.title("Relação entre Idade e Quantidade de Crédito")
plt.show()
```

```python
# Gráfico de linha: Crédito vs. Duração de crédito por Sexo

plt.figure(figsize=(10, 6))
sns.lineplot(
    data=data,
    x="Duration",
    y="Credit amount",
    hue="Sex"#,
    #ci = None
)
plt.title("Variação da Quantidade de Crédito em função da Duração, por Sexo")
plt.show()
```

```python
# Histogramas: Distribuição de Idade por Sexo (lado a lado)

plt.figure(figsize=(12, 5))

# Masculino
plt.subplot(1, 2, 1)
sns.histplot(data=data[data["Sex"] == "male"], x="Age", kde=True, color='blue')
plt.title("Distribuição de Idade - Masculino")

# Feminino
plt.subplot(1, 2, 2)
sns.histplot(data=data[data["Sex"] == "female"], x="Age", kde=True, color='pink')
plt.title("Distribuição de Idade - Feminino")

plt.tight_layout()
plt.show()
```

```python
# histograma sobreposto

plt.figure(figsize=(10, 6))
sns.histplot(
    data=data,
    x="Age",
    hue="Sex",           # separa por sexo
    kde=True,            # curva de densidade opcional
    element="step",      # estilo de linha para sobreposição mais clara
    stat="density",      # para comparar distribuições em escala
    common_norm=False,   # normaliza separadamente por grupo
    palette={"male": "blue", "female": "pink"}
)
plt.title("Distribuição de Idade por Sexo")
plt.xlabel("Idade")
plt.ylabel("Densidade")
plt.show()
```

```python
# Gráfico de barras com a média de 'Credit amount' agrupado por 'Job' e 'Sex'
plt.figure(figsize=(10, 6))
sns.barplot(
    data=data,
    x="Job",               # tipo de trabalho
    y="Credit amount",     # valor do crédito
    hue="Sex",             # por sexo (masculino/feminino)
    ci = None
)
plt.title("Média do Valor de Crédito por Tipo de Trabalho e Sexo")
plt.xlabel("Tipo de Trabalho")
plt.ylabel("Média do Valor de Crédito")
plt.xticks(rotation=45)
plt.show()
```
