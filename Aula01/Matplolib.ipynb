```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Criando gráfico básico
array = np.arange(0,10,0.1)
plt.plot(array)
plt.show()
```

```python
#Criando seno
seno = np.sin(array)
plt.plot(seno)
plt.show()
```

```python
##  Alterando tamanho da imagem, nomes dos eixos e acrescentando titulo

figura = plt.figure(figsize=(5,2))
#colocando o titulo
plt.title("Função Seno")
#plots de seno
plt.plot(seno)
plt.plot(seno*-1)
#colocando os eixos
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.show()
```

```python
## mudando cores e estilo de linha e adicionando legenda
plt.plot(seno,color='green',marker='o', label='Seno')
plt.plot(seno*-1,'r--', label='1-Seno')
plt.legend()
plt.show()
```

```python
## adicionando legendas em posições pre fixadas
plt.plot(seno,color='green',marker='o', label='Seno')
plt.plot(seno*-1,'r--', label='1-Seno')
plt.legend(frameon=False,loc='upper right')
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,3,7,1,0]) #tamanho dos eixos
```

## Bar plot with multiple bars to comparisons

```python
x1 = np.array([1,3,5,7,9])
y1 = np.array([2,3,7,1,0]) # tamanho dos eixos 1


x2 = np.array([2,4,6,8,10])
y2 = np.array([5,1,3,7,4]) # tamanho dos eixos 2
```

```python
#barplot com legendas
plt.bar(x1,y1, label = "Grup 1") #label = título das legendas
plt.bar(x2,y2, label = "Grup 2")
plt.legend()
plt.title("Bar plot")
plt.xlabel("Number")
plt.ylabel("Bar lenght")
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,3,7,1,0])
```

```python
plot_title = "Scatterplot"
x_axis = "Number"
y_axis = "Lenght"
```

## Simple scatterplot

```python
# scatterplot
plt.scatter(x,y)

plt.title(plot_title)
plt.xlabel(x_axis)
plt.ylabel(y_axis)

plt.show()
```

```python
plt.title(plot_title)
plt.xlabel(x_axis)
plt.ylabel(y_axis)

plt.plot(x,y, color = "#990000", linestyle = "--") # Ligando os pontos com uma linha tracejada
plt.show()
```

```python
# Criando os pontos com tamanhos distintos

x = np.array([1,2,3,4,5])
y = np.array([2,3,7,1,0])
z = np.array([200,25,400,3300,100]) # tamanho dos circulos
```

```python
#color = Point's color (black), marker = Marker type (.), s = marker's lenght (circle's lenght)
plt.scatter(x,y, label = "My points", color = "k", marker = ".", s = z)

plt.plot(x,y, color = "#990000", linestyle = "--") #Linha para ligar os pontos
plt.legend()
plt.show()
```

# Gráfico de Pizza

```python
cars = ['AUDI', 'BMW', 'FORD', 
        'TESLA', 'JAGUAR', 'MERCEDES'] 
  
data = [23, 17, 35, 29, 12, 41] 
fig = plt.figure(figsize =(10, 7)) 
plt.pie(data, labels = cars) 
plt.show() 
```

```python
cars = ['AUDI', 'BMW', 'FORD', 
        'TESLA', 'JAGUAR', 'MERCEDES'] 
  
data = [23, 17, 35, 29, 12, 41] 
fig = plt.figure(figsize =(10, 7)) 
plt.pie(data, labels = cars, textprops=dict(fontsize=12, color = 'w'), autopct='%.1f%%')  
plt.show() 
```

```python
def func(pct, allvals):
    #print(pct)
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return f"{pct:.1f}%\n pop: ({absolute:d} )"
```

```python
fig = plt.figure(figsize =(10, 7)) 
plt.pie(data, labels = cars, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))
plt.show()
```

```python
# explosão
explode = (0, 0.1, 0, 0, 0, 0)
```

```python
fig = plt.figure(figsize =(10, 7)) 
plt.pie(data, labels = cars, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"), explode=explode)
plt.show()
```

# Gráfico de Rosca

```python
# explosão
explode = (0.05, 0.05, 0.05, 0.05, 0.05, 0.05)
```

```python
fig = plt.figure(figsize =(10, 7)) 
plt.pie(data, labels = cars, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"), explode=explode)


# desenho de um circulo
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf() # gcf = get the current figure
 
# Adicionando o circulo ao gráfico
# gca = get the current axes
fig.gca().add_artist(centre_circle)
plt.show()
```

```python
fig = plt.figure(figsize =(10, 7)) 
plt.pie(data, labels = cars, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"), explode=explode, 
                                  pctdistance=0.85) # distancia em relacao ao raio da circunferencia)


# desenho de um circulo
centre_circle = plt.Circle((0, 0), 0.70, fc='black')
fig = plt.gcf() # pega a figura atual
 
# Adicionando o circulo ao gráfico
fig.gca().add_artist(centre_circle) #gca pega os eixos (atributos do gráfico)
plt.show()
```
